```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
#Напишите код на Python, реализующий построение графиков:
#окружности

x = []
x2 = []
y = []
y2 = []

for i in range(10000):
    r = 100
    x.append(-i)
    x2.append(i)
    y.append(np.sqrt(r ** 2 - i ** 2))
    y2.append(-np.sqrt(r ** 2 - i ** 2))
    
plt.figure(figsize = (10, 8))    
plt.plot(x, y, color = 'b')
plt.plot(x, y2, color = 'b')
plt.plot(x2, y2, color = 'b')
plt.plot(x2, y, color = 'b')
plt.axis('scaled')
plt.show()
```

```python
# эллипса

x = []
x2 = []
y = []
y2 = []

for i in range(10000):
    a = 50
    b = 80
    x.append(-i)
    x2.append(i)
    y.append(np.sqrt(b ** 2 - (b ** 2 * (i ** 2 / a ** 2))))
    y2.append(-np.sqrt(b ** 2 - (b ** 2 * (i ** 2 / a ** 2))))
       
plt.plot(x, y, color = 'r')
plt.plot(x, y2, color = 'r')
plt.plot(x2, y2, color = 'r')
plt.plot(x2, y, color = 'r')
plt.axis('scaled')
plt.show()
```

```python
# гипербола

x = []
x2 = []
y = []
y2 = []

for i in range(100):
    a = 30
    b = 50
    x.append(-i)
    x2.append(i)
    y.append(np.sqrt(b ** 2 + (b ** 2 * (i ** 2 / a ** 2))))
    y2.append(-np.sqrt(b ** 2 + (b ** 2 * (i ** 2 / a ** 2))))
       
plt.plot(x, y, color = 'g')
plt.plot(x, y2, color = 'g')
plt.plot(x2, y2, color = 'g')
plt.plot(x2, y, color = 'g')
plt.axis('scaled')
plt.show()
```
