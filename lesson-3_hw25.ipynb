```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
```

```python
#    2-  

fig = figure()
ax = Axes3D(fig)
X = np.arange(-1, 1, 1)
Y = np.arange(-1, 1, 1)

X, Y = np.meshgrid(X, Y)

Z = -10 * X + 1
Z_1 = -10 * X + 9

ax.plot_wireframe(X, Y, Z, linestyle='--', linewidth=1, colors = 'red')
ax.plot_wireframe(X, Y, Z_1, colors = 'green')

show()
```

```python
#        .

fig = figure(figsize = (10, 10))
ax = fig.add_subplot(111, projection='3d')

u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 100)
x = 30 * np.outer(np.cos(u), np.sin(v))
y = 30 * np.outer(np.sin(u), np.sin(v))
z = 30 * np.outer(np.ones(np.size(u)), np.cos(v))

ax.plot_surface(x, y, z, rstride = 2, cstride = 4, cmap = 'inferno' )

n = 10
m = 10

x = np.linspace(-10, 12, n)
y = np.linspace(-10, 12, m)

Z = np.zeros((m, n))
for i in range(m):
    for j in range(n):
        Z[i, j] = x[j]**2 + y[i]**2

X, Y = np.meshgrid(x, y)
ax.plot_surface(X, Y, Z, color = 'c')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('Z')

plt.show()
    
```
