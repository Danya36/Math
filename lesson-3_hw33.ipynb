```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Нарисуйте график функции: y(x) = k∙cos(x – a) + b для некоторых (2-3 различных) значений параметров k, a, b

x = np.linspace(-5, 10, 201)

k, a, b = np.linspace(1, 3, num = 3)
k1, a1, b1 = np.linspace(20, 35, num = 3)
k2, a2, b2 = np.linspace(32, 44, num = 3)

print(f'k={k}, a={a}, b={b}')
print(f'k1={k1}, a1={a1}, b1={b1}')
print(f'k2={k2}, a2={a2}, b2={b2}')

plt.figure()
plt.plot(x, k * np.cos(x - a) + b, label='k, a, b')
plt.plot(x, k1 * np.cos(x - a1) + b1, color='g', label='k1, a1, b1')
plt.plot(x, k2 * np.cos(x - a2) + b2, color='r', label='k2, a2, b2')

plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend(frameon=False)
plt.show()
```

```python
# Напишите код, который будет переводить полярные координаты в декартовы.

def pol2cart(rho, phi):
    x = rho * np.cos(phi)
    y = rho * np.sin(phi)
    return x, y

pol2cart (3,8) 
```

```python
# Напишите код, который будет рисовать график окружности в полярных координатах.

r = 5
rads = np.arange(0, (2 * np.pi), 0.01)
for rad in rads:
    plt.polar(rad, r, 'g.')

plt.show()
```

```python
# Решите уравнение: exp(x) + x∙(1 – y) = 1 ; y = x2 – 1

x = np.linspace(-2, 3, 201)

plt.figure()
plt.plot(x, (np.exp(x)+ x - 1)/x)
plt.plot(x, x**2 - 1)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-1,5) 
plt.grid(True)
plt.axis('scaled')
plt.show()

from scipy.optimize import fsolve

def equations(p):
    x, y = p
    return (x**2 - 1 - y, np.exp(x) + x * (1 - y) - 1)

x1, y1 =  fsolve(equations, (2, 4))
x2, y2 =  fsolve(equations, (-1, 1))
              
print (x1, y1)
print (x2, y2)
```

```python
# Решите систему уравнений: y = x2 – 1    exp(x) + x * (1 – y) > 1

x = np.linspace(-2, 3, 201)

plt.figure()
plt.plot(x, (- 1/x)+2)
plt.plot(x, x**2 - 1)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-1,5) 
plt.grid(True)

plt.show()

from scipy.optimize import fsolve

def equations(p):
    x, y = p
    return (x**2 - 1 - y, (- 1/x)+2 - y)

x1, y1 =  fsolve(equations, (3, 5))
x2, y2 =  fsolve(equations, (1, 1))
x3, y3 =  fsolve(equations, (-2, 2))
              
print (x1, y1)
print (x2, y2)
print (x3, y3)
```
